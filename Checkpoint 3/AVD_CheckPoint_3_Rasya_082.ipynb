# **Outliers**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

# **Eksplorasi Data EDA**

## **Comparison(Perbandingan)**

```python
perbandingan_region = df.groupby('region')['climate_risk_score'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

sns.barplot(x=perbandingan_region.index, y=perbandingan_region.values, palette='viridis', hue=perbandingan_region.index)
plt.title('Rata-rata Climate Risk Score Berdasarkan Region')
plt.xlabel('region')
plt.ylabel('Rata-rata Climate Risk Score')
plt.xticks(rotation=45)
plt.show()
```

**Kesimpulan:**

grafik menunjukkan bahwa Asia memiliki rata-rata Climate Risk Score tertinggi, disusul oleh Eropa dan Amerika Utara dengan nilai yang tidak jauh berbeda, sedangkan Amerika Selatan, Oseania, dan Afrika memiliki skor yang lebih rendah dan relatif mirip satu sama lain

## **Composition/Komposisi**

```python
risk_by_region = df.groupby('region')['climate_risk_score'].mean().sort_values(ascending=False)
risk_by_region_top3 = risk_by_region.head(3)

plt.figure(figsize=(10, 8))
risk_by_region_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Region dengan Climate Risk Score Tertinggi')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

**Kesimpulan:**

diagram pie menunjukkan bahwa tiga region dengan Climate Risk Score tertinggi—Asia, Afrika, dan Amerika Utara—memiliki proporsi yang hampir sama, masing-masing sekitar 33%, sehingga kontribusi ketiganya relatif seimbang tanpa perbedaan yang signifikan

## **Distribution/Distribusi**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['climate_risk_score'], bins=20, kde=True)
plt.title('Distribusi Climate Risk Score')
plt.xlabel('Climate Risk Score')
plt.ylabel('Frekuensi')
plt.show()
```

**Kesimpulan:**

histogram menunjukkan bahwa Climate Risk Score tersebar cukup merata pada rentang sekitar 20 hingga 95, tanpa lonjakan yang terlalu ekstrem di satu titik tertentu, sehingga distribusinya relatif stabil dengan kecenderungan peningkatan frekuensi pada nilai menengah hingga tinggi

## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['co2_emissions_mt', 'climate_risk_score']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara CO2 Emissions dan Climate Risk Score')
plt.show()
```

**Kesimpulan:**

heatmap menunjukkan bahwa korelasi antara CO2 Emissions dan Climate Risk Score sangat lemah (sekitar 0,01), sehingga hampir tidak ada hubungan linear yang signifikan antara kedua variabel tersebut, artinya tingginya emisi CO₂ tidak secara langsung berkorelasi kuat dengan tinggi rendahnya Climate Risk Score dalam data ini
